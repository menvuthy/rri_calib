```python
import os
import glob
import spotpy
import subprocess
import numpy as np
import pandas as pd
from scipy import interpolate
import matplotlib.pyplot as plt
import proplot
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import mean_squared_error, r2_score

# Define the RRI_Model class
class RRI_Model:
    def __init__(self, calibration_file):
        
        # Initialize parameters from the inputCalibration.txt file
        self.params = self.read_calibration_file(calibration_file)
        
        # Automatically retrieve station files
        self.station_files = [os.path.basename(file_path) for file_path in glob.glob('disc_obs/*.txt')]
        

    def read_calibration_file(self, file_path):
        params_list = []
        
        # Define a list of parameter names in the order of the file
        param_names = ['ns_river', 'ns_slope', 'soildepth', 'gammaa', 'kv', 'Sf', 'ksg', 'gammag', 'kg0', 
                      'width_param_c', 'width_param_s', 'depth_param_c', 'depth_param_s']
        
        with open(file_path, 'r') as file:
            lines = file.readlines()
            
            # Iterate over lines and param_names together
            for i, line in enumerate(lines[1:]):
                # Split the line by '#' to remove the comment, keep the part before '#'
                line = line.split('#')[0].strip()

                # Now split by comma and strip whitespace
                low, high = [float(x.strip()) for x in line.split(',')]

                # Create a Uniform parameter with the corresponding name, low, and high values
                step_value = 0.001
                param = spotpy.parameter.Uniform(param_names[i], low=low, high=high, step=step_value)
                params_list.append(param)
        
        return params_list
    
    # Function to generate parameters
    def parameters(self):
        params = spotpy.parameter.generate(self.params)
        return params
    
    def update_rri_input_file(self, vector):
        with open('RRI_Input.txt', 'r') as file:
            lines = file.readlines()

        lines[17] = f"{vector[0]}     # ns_river\n"
        lines[20] = f"{vector[1]}     # ns_slope\n"
        lines[21] = f"{vector[2]}     # soildepth\n"
        lines[22] = f"{vector[3]}     # gammaa\n"
        lines[24] = f"{vector[4]}     # kv\n"
        lines[25] = f"{vector[5]}     # Sf\n"
        lines[31] = f"{vector[6]}     # ksg\n"
        lines[32] = f"{vector[7]}     # gammag\n"
        lines[33] = f"{vector[8]}     # kg0\n"
        lines[38] = f"{vector[9]}     # width_param_c\n"
        lines[39] = f"{vector[10]}     # width_param_s\n"
        lines[40] = f"{vector[11]}     # depth_param_c\n"
        lines[41] = f"{vector[12]}     # depth_param_s\n"

        with open('RRI_Input.txt', 'w') as file:
            file.writelines(lines)
        
        # update ns river at CS section
        with open('riv/section/ns_river.txt', 'r') as file:
            lines = file.readlines()
            
        lines[0] = f"{vector[0]}"
        
        with open('riv/section/ns_river.txt', 'w') as file:
            file.writelines(lines)
    
    def updateSection(self):
        # Define project directory path
        original_path = os.getcwd()

        # Step 2: Extend the path to 'riv/section'
        extended_path = os.path.join(original_path, 'riv', 'section')
        os.chdir(extended_path)

        # Execute creating section
        !python createSection.py

        # Return to project path
        os.chdir(original_path)

            
    def run_rri(self):        
        subprocess.run(['0_rri_1_4_2_6.exe'])
        
    def calcHydro(self):        
        subprocess.run(['calcHydro.exe'])
        
    def read_sim(self):
        sim_discharge_list = []
        
        for station_file in self.station_files:
            sim_disc_df = pd.read_csv(f'disc_sim/{station_file}', delimiter=r"\s+", header=None)

            # Extract time and discharge values
            sim_time, sim_discharge = sim_disc_df[0], sim_disc_df[1]

            # Generate simulated values for the observed time steps
            sim_discharge_list.append(sim_discharge)

        return sim_discharge_list

    def plotHydro(self, avg_rmse):
        obs_disc_path = glob.glob('disc_obs/*.txt')
        sim_disc_path = glob.glob('disc_sim/*.txt')
        sta_name = [os.path.basename(filepath)[:-4] for filepath in sim_disc_path]
        
        # Parent directory where the folders will be created
        parent_dir = 'graph'

        outpath = []
        # Create the directories
        for folder in sta_name:
            # Create the full path by joining the parent directory and the folder name
            folder_path = os.path.join(parent_dir, folder)
            outpath.append(folder_path)
            # Check if the directory already exists, if not, create it
            if not os.path.exists(folder_path):
                os.makedirs(folder_path)

        obs_list = []
        for path in obs_disc_path:
            obs = pd.read_csv(path, delimiter=r"\s+", header=None)
            obs_list.append(obs)

        sim_list = []
        for path in sim_disc_path:
            sim = pd.read_csv(path, delimiter=r"\s+", header=None)
            sim_list.append(sim)

        for i in range(len(obs_list)):
            obs_disc_df = obs_list[i]
            sim_disc_df = sim_list[i]
            # Extract the time and discharge values
            obs_time, obs_discharge = obs_disc_df[0], obs_disc_df[1]
            sim_time, sim_discharge = sim_disc_df[0], sim_disc_df[1]

            # Interpolation (using cubic spline for smooth curves)
            fine_time = np.linspace(obs_time.min(), obs_time.max(), 1000)  # Create a finer time grid

            # Create cubic spline functions for observed and simulated discharges
            obs_spline = interpolate.CubicSpline(sim_time, obs_discharge)
            sim_spline = interpolate.CubicSpline(sim_time, sim_discharge)

            # Calculate R^2 and RMSE
            r2 = r2_score(obs_discharge[:], sim_discharge[:])
            rmse = np.sqrt(mean_squared_error(obs_discharge[:], sim_discharge[:]))

            # Plot the observed and simulated discharge with smooth curves
            plt.figure(figsize=(7, 3))

            # Plot observed discharge (interpolated)
            plt.plot(fine_time, obs_spline(fine_time), label='Observation', color='black', linestyle='-', lw=1.5)

            # Plot simulated discharge (interpolated)
            plt.plot(fine_time, sim_spline(fine_time), label='Simulation', color='green', linestyle='-', lw=1.5)

            # Add labels and title
            plt.xlabel('Number of Output', fontsize=10)
            plt.ylabel('Discharge [m³/s]', fontsize=10)
            plt.title('Magat River', fontweight='bold', fontsize=10)
            plt.title(f'{sta_name[i]}\nR²: {r2:.2f}, RMSE: {rmse:.2f}', fontsize=10)

            # Add legend
            plt.legend(fontsize=10, ncol=2)

            # Show the plot
            plt.grid(False)
            plt.tight_layout()
            # Save plot with iteration number
            plt.savefig(f'{outpath[i]}/{sta_name[i]}_avgrmse_{avg_rmse}.png', dpi=100)
            plt.close()
    
    def simulation(self, vector):
        self.update_rri_input_file(vector)
        self.updateSection()
        self.run_rri()
        self.calcHydro()

        return self.read_sim()
    
    def evaluation(self):
        obs_discharge_list = []

        for station_file in self.station_files:
            obs_disc_df = pd.read_csv(f'disc_obs/{station_file}', delimiter=r"\s+", header=None)
            obs_discharge_list.append(obs_disc_df[1])
            
        return obs_discharge_list
    
    def objectivefunction(self, simulation, evaluation):
        rmse_list = []
        mae_list = []
        r2_list = []

        for sim_data, obs_data in zip(simulation, evaluation):
            # Calculate RMSE
            rmse = np.sqrt(np.mean((obs_data[:] - sim_data[:]) ** 2))
            rmse_list.append(rmse)

            # Calculate MAE
            mae = np.mean(np.abs(obs_data - sim_data))
            mae_list.append(mae)

            # Calculate R²
            obs_mean = np.mean(obs_data)
            ss_total = np.sum((obs_data - obs_mean) ** 2)
            ss_residual = np.sum((obs_data - sim_data) ** 2)
            r2 = 1 - (ss_residual / ss_total)
            r2_list.append(r2)

        # Average metrics across all stations
        avg_rmse = np.mean(rmse_list)
        avg_mae = np.mean(mae_list)
        avg_r2 = np.mean(r2_list)
        
        self.plotHydro(avg_rmse)
        
        # Return a tuple of metrics (you can choose which metrics to include)
        return [avg_rmse]  # Returning multiple values
```

```python
# Initialize Model:
calibration_file = 'inputCalibration.txt'
rri_model = RRI_Model(calibration_file)
sampler = spotpy.algorithms.sceua(rri_model, dbname='RRI_Calibration', dbformat='csv')

# Run Calibration with the SCE-UA algorithm
with open(calibration_file, 'r') as file:
    lines = file.readlines()
    iteration = int(lines[0].split('#')[0].strip())

print('Number of Iteration:', iteration)
rri_model.params
```

```python
# Start Calibration
sampler.sample(iteration)
```
