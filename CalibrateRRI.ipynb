# Auto‑Calibration for Rainfall-Runoff Inundation (RRI) Model

This notebook performs **automatic calibration of the Rainfall–Runoff–Inundation (RRI) model**
using the **Shuffled Complex Evolution – University of Arizona (SCE‑UA)** global optimization
algorithm implemented in **SPOTPY**.

## Overview

The purpose of this notebook is to calibrate key RRI model parameters by minimizing the
difference between **observed** and **simulated** discharge time series.

The calibration framework:
- Couples the RRI model with the SPOTPY optimization library
- Uses the SCE‑UA algorithm for global parameter search
- Evaluates model performance using statistical objective functions (e.g. RMSE, MAE, R²)

This workflow is intended to reduce manual trial‑and‑error and improve the reproducibility
of RRI model calibration.

## Calibration Workflow

The calibration process follows these steps:

1. Initialize the RRI calibration model (`CalibrateRRI`)
2. Read calibration settings and parameter ranges
3. Load observed discharge data
4. Run repeated RRI simulations with different parameter sets
5. Evaluate model performance using an objective function
6. Search for optimal parameters using the SCE‑UA algorithm
7. Save calibration results and logs

## Input Data

The following inputs are required before running the calibration:

- Observed discharge data (CSV format) with filename starting from `disc_ + station name`, and must be saved in `rri_calib/obs` folder
- RRI model input files and executable
- Parameter ranges defined in `ParameterSetting.xlsx`
- Number of calibration iterations
- Selected objective function (e.g. `RMSE` or `MAE`)

Please ensure that all input files are correctly configured before execution.

## Notes / Disclaimer 

- Calibration results depend strongly on the quality of observed data and the
  selected parameter ranges.
- It is recommended to verify calibrated parameters against physical
  and hydrological plausibility.
- Long calibration runs are advised to ensure stable convergence of the SCE‑UA algorithm.

#### <font color='green'>**Import Modules**</font>

```python
import spotpy
from rri_calib import CalibrateRRI
```

#### <font color='green'>**Start Auto-Calibration**</font>

```python
# --------------------------------------------------
# Initialize RRI calibration model
# --------------------------------------------------
RRI_model = CalibrateRRI()

# --------------------------------------------------
# Load calibration parameters and settings
# --------------------------------------------------
iteration = RRI_model.read_iteration()
log_sheet = RRI_model.calibration_log()

# --------------------------------------------------
# Apply SCE-UA optimization algorithm
# --------------------------------------------------
sampler = spotpy.algorithms.sceua(
    RRI_model,
    dbname=log_sheet,
    dbformat="csv"
)

# --------------------------------------------------
# Start auto-calibration
# --------------------------------------------------
sampler.sample(iteration)
```
